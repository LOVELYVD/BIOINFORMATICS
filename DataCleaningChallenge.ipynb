```python
import pandas as pd
data=pd.read_csv('data_cleaning_challenge.csv')
data=data.drop(columns=['Unnamed: 9','Unnamed: 10'])
df=data.rename({'Row Type':'FirstName','Iter Number':'LastName','Power1':'Date'},axis=1)
df=df[df['FirstName'].notna()]
column_values=[]
counter=0
for i in df['FirstName']:
  if 'first name' in i:
   counter+=1
  column_values.append(counter)
df1=df
df1['counters']=column_values
df1

```

```python
df=df1[df1['FirstName']!='Row Type']
df
```

```python
df2=df[df['FirstName'].str.contains('first name')]
df2.drop(columns=['Speed1','Speed2','Electricity','Effort','Weight','Torque'],inplace=True)
df2.FirstName,df2.LastName,df2.Date=df2.FirstName.str[12:],df2.LastName.str[11:],df2.Date.str[6:]
df2
```

```python
df3=df[~df['FirstName'].str.contains('first name')]
df3
```

```python
df=pd.merge(left=df2,right=df3,how='inner',on='counters')
df
```
